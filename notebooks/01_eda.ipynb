# Relationship Discovery

# 1.Soil Moisture vs Irrigation Required

Business Question

Do farms with low soil moisture require irrigation?

```python

colors = ['lightgreen', 'salmon']

sns.boxplot(data=df,x='irrigation_required',y='soil_moisture',palette=colors)
plt.xlabel('Irrigation_required')
plt.ylabel('soil_moisture')
plt.show()
```

# Interpretation  
The boxplot reveals a strong relationship between soil moisture and irrigation requirement. Farms that require irrigation have considerably lower soil moisture levels, with a median moisture level around 20%. In contrast, farms that do not require irrigation exhibit much higher moisture levels, with a median around 45%.

This indicates that soil moisture is a key factor in irrigation decision-making. As soil moisture decreases, the likelihood of irrigation increases. Therefore, soil moisture can be considered one of the most important predictors for the Irrigation Required target variable.

Conclusion:
There is a strong negative relationship between soil moisture and irrigation requirement. Lower soil moisture levels are associated with a higher need for irrigation, making soil moisture one of the most influential variables in the Smart Irrigation Prediction System.

2. Rainfall vs Irrigation Required

Business Question

Does irrigation decrease when rainfall increases?

```python
sns.boxplot(data=df,x='irrigation_required',y='rainfall')
plt.xlabel('Irrigation_required')
plt.ylabel('rainfall')
plt.show()
```

# Interpretation
The average rainfall for farms requiring irrigation is slightly higher than for farms that do not require irrigation. This suggests that rainfall alone is not the primary factor influencing irrigation decisions in this dataset.

Other variables such as soil moisture, crop type, temperature, evapotranspiration, and soil conditions may have a stronger impact on irrigation requirements.

# Business insights
1. Rainfall does not show a strong negative relationship with irrigation requirement.
2. Higher rainfall does not necessarily eliminate the need for irrigation.
3. Soil moisture appears to be a more influential factor than rainfall.
4. Irrigation decisions likely depend on multiple environmental and soil factors rather than rainfall alone.

# Conclusion:
Rainfall vs Irrigation Required:

The analysis indicates that rainfall alone is not a strong predictor of irrigation requirement. Farms requiring irrigation exhibit slightly higher average rainfall than farms not requiring irrigation. Therefore, irrigation decisions are likely influenced by a combination of factors, with soil moisture being a more significant driver.

# 3. Temperature vs Water Quantity Required

Business Question

Do hotter conditions require more water?

```python
sns.scatterplot(data=df,x='temperature',y='water_quantity_required_liters',hue='irrigation_required')
plt.xlabel('Temperature (°C)')
plt.ylabel('Water Quantity Required (liters)')
plt.title('Temperature vs Water Quantity Required')     
plt.show()      
```

#Interpretation

The scatter plot indicates a positive relationship between temperature and water quantity required. As temperature increases, the amount of water required by crops also tends to increase.

Higher temperatures accelerate evaporation and plant transpiration, resulting in greater water demand. Consequently, farms operating under hotter environmental conditions require more irrigation water to maintain optimal crop growth.

Business Insights
1. Water requirement increases with rising temperature.
2. Temperature is an important predictor for estimating irrigation water quantity.
3. Hot weather conditions contribute to higher crop water demand due to increased evaporation and transpiration.
4. Smart irrigation systems should consider temperature when calculating irrigation schedules and water allocation.

Conclusion:
There is a moderate to strong positive relationship between temperature and water quantity required. Crops grown under higher temperature conditions generally require more water, making temperature one of the key features for the Water Quantity Required prediction model.

# **Correlation Study**

```python
# separate numerical and categorical feature

num_features =df.select_dtypes(include=['int64','float64']).columns
cat_features= df.select_dtypes(include=['object']).columns
```

#water quantity required correlation

```python
corr_target = df.corr(numeric_only=True)['water_quantity_required_liters'].sort_values(ascending=False).head(10)
print(corr_target)

```

```python
#Heatmap of Important Features + Target

selected_features=['temperature','evapotranspiration','humidity', 'soil_moisture','water_quantity_required_liters']
df_subset= df[selected_features]

corr_matrix = df[selected_features].corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap - Water Quantity Required')
plt.show()
```

# Interpretation:
The correlation analysis was conducted to identify the factors influencing water quantity requirements for irrigation.

The results indicate that soil moisture is the most important factor affecting water demand, showing a strong negative correlation (-0.773). This suggests that as soil moisture increases, the amount of irrigation water required decreases significantly.

Temperature (0.305) and evapotranspiration (0.299) exhibit moderate positive correlations with water quantity requirements. This indicates that crops require more water under hotter conditions and when water loss through evaporation and transpiration increases.

Humidity demonstrates a negligible correlation (0.009), indicating that it has little direct influence on water quantity requirements in this dataset.
